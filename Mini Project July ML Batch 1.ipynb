```python
third = a.sort_values(['budget'])
third_lowest=third.iloc[2]['original_title']
thi = a.sort_values(['budget'])
third_highest=thi.iloc[-3]['original_title']
third_lowest, third_highest 
```

```python
movies=a[a['release_year'].isin([2000, 2001, 2002, 2003, 2004, 2005])]
words=0
for i in range(movies.shape[0]):
    words_l = movies['original_title'].values[i].split(' ')
    words = words + len(words_l)
    
avg_words = words/movies.shape[0]
avg_words
```
